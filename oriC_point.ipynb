```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#https://bioinformaticsalgorithms.com/data/realdatasets/Replication/Vibrio_cholerae.txt
```

```python
df = pd.read_table('/content/Vibrio_cholerae.txt', header=None, names=['sequence'])
text = df['sequence'][0]
```

```python

list_diff = []
c = 0
g = 0

for i in range(0, len(text)-1):
  if text[i] == 'C':
    c+=1
  elif text[i] == 'G':
    g+=1
  list_diff.append(g-c)


plt.plot(list_diff)
plt.scatter(list_diff.index(max(list_diff)), max(list_diff), color='red')
plt.scatter(list_diff.index(min(list_diff)), min(list_diff), color='green')
plt.xlabel('Position in Nucleotide')
plt.ylabel('Difference (G - C) Statistics')
plt.title('Ter is Red /Ori is Green')
plt.show()
```
