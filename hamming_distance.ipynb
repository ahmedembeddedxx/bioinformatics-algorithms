```python
import pandas as pd
```

```python
df = pd.read_table('genome1.txt', header=None, names=['sequence'])
text1 = df['sequence'][0]

df = pd.read_table('genome2.txt', header=None, names=['sequence'])
text2 = df['sequence'][0]
```

```python
print(f'Genome 1: {text1}')
print(f'Genome 2: {text2}')
```

```python
def hamming_distance(gen1: str, gen2: str)->int:
  l = 0
  if(len(gen1) > len(gen2)):
    l = len(gen2)
  else:
    l = len(gen1)

  ctr = 0
  for i in range(0, l-1):
    if(gen1[i] != gen2[i]):
      ctr+=1

  return [ctr, l]
```

```python
print(f'Hamming Distance b/w Genome1 & Genome2 is {hamming_distance(text1, text2)[0]}, with cumulative length {hamming_distance(text1, text2)[1]}')
```
